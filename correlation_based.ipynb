```
import numpy as np
import pandas as pd
```

```
data = pd.read_csv('/content/gdrive/My Drive/Book Recommendation/listing.csv', encoding = 'latin-1')
books = pd.read_csv('/content/gdrive/My Drive/Book Recommendation/books.csv', encoding = 'latin-1')
```

```
data.head()
```

```
books.head()
```

```
rating = pd.DataFrame(books.groupby('book_id')['no_of_ratings'].mean())
rating.head()
```

```
rating.describe()
```

```
rating.sort_values('no_of_ratings', ascending = False).head()
```

```
rating_pivot = pd.pivot_table(data = books, values = 'user_rating', index = 'user_id', columns = 'book_id')
rating_pivot.head()
```

```
corr = rating_pivot.corr()
corr.head(10)
```

```
id = int(input('Enter book id '))
one_rating = rating_pivot[id]
one_rating[one_rating >= 0]
```

```
similar_rating = rating_pivot.corrwith(one_rating)
corr_rating = pd.DataFrame(similar_rating, columns = ['PearsonR'])
corr_rating.dropna(inplace = True)
corr_rating.head()
```

```
one_rating = corr_rating.join(rating)
one_rating.sort_values('PearsonR', ascending = False).head(10)
```
